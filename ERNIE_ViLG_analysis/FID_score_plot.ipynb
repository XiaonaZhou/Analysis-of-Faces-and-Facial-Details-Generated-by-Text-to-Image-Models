```python
# Calculate the average
Real_mean = np.mean(Real_FID)
Stable_Diff_mean = np.mean(Stable_Diff_FID)
DALL_E_2_mean = np.mean(DALL_E_2_FID)
Mid_mean = np.mean(Mid_FID)
ERNIE_mean = np.mean(ERNIE_FID)
```

```python
Real_mean
Stable_Diff_mean
DALL_E_2_mean
Mid_mean
ERNIE_mean
```

```python
# Calculate the standard deviation
Real_std = np.std(Real_FID)
Stable_Diff_std = np.std(Stable_Diff_FID)
DALL_E_2_std = np.std(DALL_E_2_FID)
Mid_std = np.std(Mid_FID)
ERNIE_std = np.std(ERNIE_FID)
```

```python
Real_std
Stable_Diff_std
DALL_E_2_std
Mid_std
ERNIE_std
```

```python
# Create lists for the plot
model_names = ['Real Images',  'Stable Diffusion', 'DALL-E 2', 'Midjourney', 'ERNIE-ViLG']
x_pos = np.arange(len(model_names))
FID_means = [Real_mean, Stable_Diff_mean, DALL_E_2_mean, Mid_mean, ERNIE_mean]
error = [Real_std, Stable_Diff_std, DALL_E_2_std, Mid_std, ERNIE_std]
```

```python
# Build the plot
fig, ax = plt.subplots()
ax.bar(x_pos, FID_means, yerr=error, align='center', alpha=0.5, ecolor='black', capsize=10)
ax.set_ylabel('FID')
ax.set_xticks(x_pos)
ax.set_xticklabels(model_names)
ax.set_title('5k Images')
ax.yaxis.grid(True)
plt.ylim((0,100))

# Save the figure and show
plt.tight_layout()
plt.savefig('FID_on_faces_reproduced.png')
plt.show()
```
